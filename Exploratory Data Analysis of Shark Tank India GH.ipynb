# Present of Shark for Episode

```python
e = shark['peyush_present'].value_counts()[1]
print("Shark Peyush Bansal Present :", e ,"episode")
```

```python
f = shark['ghazal_present'].value_counts()[1]
print("Shark Ghazal Alagh Present :", f ,"episode")
```

```python
g = shark['aman_present'].value_counts()[1]
print("Shark Aman Gupta Present :", g ,"episode")
```

```python
aman_present = len(shark[shark.aman_present == 1])
anupam_present = len(shark[shark.anupam_present == 1])
ashneer_present = len(shark[shark.ashneer_present == 1])
ghazal_present = len(shark[shark.ghazal_present == 1])
namita_present = len(shark[shark.namita_present == 1])
peyush_present = len(shark[shark.peyush_present == 1])
vineeta_present = len(shark[shark.vineeta_present == 1])
```

```python
fig = plt.figure(figsize = (10, 5))
present = [aman_present, anupam_present, ashneer_present, ghazal_present,
namita_present, peyush_present, vineeta_present]
names = ['Aman','Anupam', 'Ashneer', 'Ghazal', 'Namita', 'Peyush', 'Vineeta']
plt.bar(names, present, color ='b')
plt.xlabel("Shark's  Name")
plt.ylabel("Shark's  Present")
plt.title("Shark's Present as per Episode")
plt.show()
```

In the given Dataset, Anupam present was Highest and Ghazal Present was very low.


# Percentage of Sharks Present

```python
plt.pie(present, labels = names, radius = 2.0, autopct = '%0.1f%%')
plt.suptitle('Percentage of Sharks Present',
y = 1.15, color = 'red')
plt.show()
```

The present percentage of sharks :

Aman Gupta ---> 16.4%

Ashneer Grover ---> 16.4%

Vineeta Singh ---> 11.0%

Peyush Bansal ---> 14.7%

Namita Thapar ---> 17.7% 

Anupam Mittal ---> 19.5% 

Ghazal Alagh ---> 4.3%
